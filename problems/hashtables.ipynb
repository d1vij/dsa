````python
# hash table with hash collision handled using linear probing

# in linear probing, if two hashed indices collide, then the next the next node empty node is set the value
# worst case lookup is o(n)
# if end of array is reached, then lookup starts from the start and if we reach the initial starting index, errror is raised that no empty space remains

from typing import Any


class LinearProbed_HashTable:
    def __init__(self, size: int) -> None:
        self.size = size
        self.__array: list[None | tuple[str, Any]] = [None for _ in range(self.size)]

    def __str__(self):
        return str(self.__array)
    def hash(self, key):
        return sum(ord(c) for c in key) % self.size

    def __setitem__(self, key, value):
        hash_index = self.hash(key)

        if(self.__array[hash_index] is None):
            # particular index is empty
            self.__array[hash_index] = (key, value)
            return

        for i in range(hash_index, self.size):
            # linear probe from hash_index to end of array
            if(self.__array[i] is None):
                self.__array[i] = (key, value)
                return

        for i in range(hash_index):
            # linear probe from start to hash_index

            if(self.__array[i] is None):
                self.__array[i] = (key, value)
                return

        # no empty space
        raise KeyError("Cannot find empty space in hashmap !")

    def __getitem__(self, key):
        hash_index = self.hash(key)

        for i in range(hash_index, self.size):
            # ideally the to be found key should be ahead of hash_index
            if (self.__array[hash_index] is None):
                continue

            if (self.__array[i][1]==key):
                return self.__array[i][1]
        for i in range(hash_index):
            if self.__array[hash_index] is None:
                continue

            if self.__array[i][1] == key:
                return self.__array[i][1]
        
        raise KeyError("Value for key not found !") # should not occur; js for safety
    
lpht = LinearProbed_HashTable(10)
print(lpht)
lpht['march 6'] = 1
lpht['march 10'] = 2
lpht['march 17'] = 3
print(lpht)
````
